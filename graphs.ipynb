```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values for the two segments of the curve
x1 = np.linspace(0.01, 17.4, 500)  # Avoid starting from 0 to allow log scale
x2 = np.linspace(17.4, 900, 500)  # You can adjust the upper limit as needed

# Define the y values based on the given equations
y1 = 15.7/17.4 * x1
y2 = np.full_like(x2, 15.7)

# Plot the curve
plt.figure()
plt.figure(figsize=(10, 6))
plt.plot(x1, y1)
plt.plot(x2, y2)

plt.scatter(17.4, 15.7, color='red', zorder=20, label='V100 Limit (12.54, 19.5)')


# VGG-11
plt.scatter(25.253, .941, color='orange', zorder=5, label='VGG-11 - (25.253, .941)')

# VGG-16
plt.scatter(40.33, 1.856, color='violet', zorder=5, label='VGG-16 - (40.33, 1.856)')

# VGG-19
plt.scatter(45.674, 1.974, color='blue', zorder=5, label='VGG-19 - (45.674, 1.974)')


# Set log scale for both axes
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.xlabel('AI, Flops/bytes')
plt.ylabel('TFlops/sec')
plt.title('Roofline for V100')
plt.legend()

# Show the plot
plt.show()
```
